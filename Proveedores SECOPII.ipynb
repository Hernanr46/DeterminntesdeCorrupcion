```python
#Importar Base

client = Socrata("www.datos.gov.co", None)
Proveedores = client.get("qmzu-gj57",
                content_type="csv", limit=1116919)
Proveedores
```

```python
#Titulos
columns = Proveedores[0]
del(Proveedores[0])
Proveedores = pd.DataFrame(Proveedores)
```

```python
#DataFrame
Proveedores.columns = columns
Proveedores
```

```python
#Se crea base nueva filtrando el estado del contrato sin Borrador

Proveedores = Proveedores[Proveedores['nit'] !='No Definido'].copy()
```

```python
#Se crea base nueva filtrando el estado del contrato sin Borrador

Proveedores = Proveedores[Proveedores['nit'] !='N.N.'].copy()
```

```python
#Se crea base nueva filtrando el estado del contrato sin Borrador

Proveedores = Proveedores[Proveedores['nit'] !='NNNNNN'].copy()
Proveedores = Proveedores[Proveedores['nit'] !='Noadjudicatario'].copy()
Proveedores = Proveedores[Proveedores['nit'] !='No aplica'].copy()
Proveedores = Proveedores[Proveedores['nit'] !='no aplica'].copy()
Proveedores = Proveedores[Proveedores['nit'] !='NOAPLICA'].copy()
Proveedores = Proveedores[Proveedores['nit'] !='NO APLICA'].copy()
Proveedores = Proveedores[Proveedores['nit'] !='NO APLICA-1'].copy()
Proveedores = Proveedores[Proveedores['nit'] !='NO APLICA 2'].copy()
```

```python
Proveedores = Proveedores.drop(['fecha_creacion','pais','codigo'], axis=1)
```
